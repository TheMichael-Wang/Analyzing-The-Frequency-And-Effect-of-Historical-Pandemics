### Data Loading
    Here I imported the code from two csv files that are relevant to the analysis that provide data on various outbreaks and pandemics throughout history, which are called df_1.csv and df_16.csv respectively. 

```python
import pandas as pd
import re

df = pd.read_csv("df_1.csv")
df3 = pd.read_csv("df_16.csv")
```

### Analysis
    Here I calculated how frequently major pandemics have occured throughout human history, and plotted a graph showing how many pandemics there were in each era.

```python
dates = df['Date']
start = []
for i in dates:
     start.append(re.split(r'\D+',i)[0])
earliest = min(start)
start = list(map(int,start))
earliest = (min(start))

frequency = round((2022-earliest)/19)

print("There were 19 major pandemics with death tolls of over 1 million in recorded history, equating to roughly one every",frequency, "years.")

#outbreaks each era
counts = []
def count(row):
    c = 0
    for i in row:
        if i == ")":
            c+=1
    return c

list1 = df3.values.tolist()
for i in list1:
    temp = ''.join(map(str, i))
    counts.append(temp.count(')'))
    
graphdf = pd.DataFrame({'era':['Ancient','Post-classical','Early modern','16th century','17th century','18th century','Modern','19th century','20th century','21st century'],'y':counts})
graph = graphdf.plot.bar(x='era', y='y', rot=90)
```

### Conclusions
    I first discovered that large-scale pandemics with death tolls of over one million people were quite rare, with only 19 in recorded history, which equates to roughly one every 98 years. However, later data showed a different story. Although pandemics are quite rare over the entirety of human history, the frequency of pandemics appearing has increased drastically from ancient times. As the graph shows, there were only 24 major outbreaks in pre modern times, whereas there were a whopping 71 in modern times. Evidently, pandemics are becoming more frequent than ever before, with no signs of stopping any time soon, and will only continue becoming an even more pressing issue in the future. Therefore there must be more attention and funding put towards public healthcare and preventing future pandemics that are sure to come.
